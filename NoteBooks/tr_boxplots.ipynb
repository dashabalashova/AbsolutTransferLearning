```python
import pandas as pd

df = pd.read_csv('../Results/transf_metr_m4.tsv', sep='\t')
df['type_lr'] = df.apply(lambda x:x.type+'_'+str(x.lr), axis=1)
```

```python
print(df[df.type=='CNNModel'].rocauc_val.max())
print(df[df.type=='CNNModel_transf'].rocauc_val.max())
print(df[df.type=='CNNModel_freezing'].rocauc_val.max())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set(rc={'figure.figsize':(11.7,8.27)})
sns.set_theme(style="ticks", font_scale=1.25)

ax = sns.lineplot(data=df, x="epoch", y="rocauc_val", hue="type", style="lr", estimator=None)
ax.set_xlabel("Number of epochs", fontsize=18)
ax.set_ylabel("ROC AUC",fontsize=18)
```
